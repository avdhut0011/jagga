```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
at = pd.read_csv('C:/Users/Shree/Downloads/temperatures.csv')
```

```python
at.head()
```

```python
x = at['YEAR']   #Input data
y = at['ANNUAL']   #Output data
```

```python
plt.figure(figsize=(10,5))     #its dimension of our plot
plt.title('temperature plot of india')     #gives title name of plot
plt.xlabel('Year')                       #gives x-axis name
plt.ylabel('Annual avg. tempurature')      #gives y-axis name
plt.scatter(x,y)                          #display the scatter plot
```

```python
x.shape
```

```python
x = x.values    #using values function we can convert data into array
```

```python
x      #display the array
```

```python
x = x.reshape(117,1)    # we can reshape the array into various dimensions
```

```python
x.shape    
```

```python
from sklearn.linear_model import LinearRegression    #we import LinearRegression method from linear_model subpackage of sklearn library
```

```python
regressor = LinearRegression()    # here we can create an regressor object which calls LinearRegression() constructor
```

```python
#for training the dataset we put the parameters input and output into fit method , it gives trained model
regressor.fit(x,y)
```

```python
regressor.coef_     #it gives slope of object means y=mx+c ,here it is m.
```

```python
regressor.intercept_    # it gives model parameters or y-intercept means when regression line intercept the y-axis that point it gives us
```

```python
regressor.predict([[2024]])  #after creating model we can predict the future tempurature value using predict method
```

```python
predicted = regressor.predict(x)    #we can find predicted value using predict method with input parameter x 
```

```python
np.mean(abs(y-predicted))    # it gives mean absolute error by substracting predicted value from output data(y).
```

```python
from sklearn.metrics import mean_absolute_error    # it also goves mean_absolute_error directly using sklearn library method
mean_absolute_error(y,predicted)
```

```python
from sklearn.metrics import mean_squared_error    # it also goves mean_squared_error directly using sklearn library method
mean_squared_error(y,predicted)
```

```python
from sklearn.metrics import r2_score   # it also goves r2_score directly using sklearn library method
r2_score(y,predicted)
```

```python
plt.figure(figsize=(10,5))     #its dimension of our plot
plt.title('temperature plot of india')     #gives title name of plot
plt.xlabel('Year')                       #gives x-axis name
plt.ylabel('Annual avg. tempurature')      #gives y-axis name
plt.scatter(x,y,label = 'actual')         # plot the scatter graph with label name as actual
plt.plot(x,predicted,label = 'predicted' ,color='r')    # plot the graph with label name as predicted with color red
plt.legend()                            # it gives corner box of graph elements it is not essential.
```

```python
# we can also directly plot the graph without above procedure using seaborn 
sns.regplot(x='YEAR', y='ANNUAL', data=at)
```
